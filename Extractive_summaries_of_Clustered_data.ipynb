## Function to display Keywords as a word cloud and display Summaries

```
from matplotlib import gridspec
import matplotlib.patheffects as path_effects
import textwrap

def display_keywords(a,b,c):

  if b != None:  
    field = new_df.tag.values[a]
    wordcloud = WordCloud( background_color="white", max_words=20).generate(str(field))
    field1 = new_df.tag.values[b]
    wordcloud1 = WordCloud( background_color="white", max_words=20).generate(str(field1))
    field2 = new_df.tag.values[c]
    wordcloud2 = WordCloud( background_color="white", max_words=20).generate(str(field2))
    text1 = field1 = new_df.extractive_summary.values[0]
    f = plt.figure(figsize=(20,10))
    ax1 = f.add_subplot(131)
    ax2 = f.add_subplot(132)
    ax3 = f.add_subplot(133)
    ax1.imshow(wordcloud)
    ax2.imshow(wordcloud1)
    ax3.imshow(wordcloud2)
  else:
    field = new_df.tag.values[a]
    wordcloud = WordCloud( background_color="white", max_words=20).generate(str(field))
    f = plt.figure(figsize=(10,5))
    ax1 = f.add_subplot(111)
    ax1.imshow(wordcloud)

  return

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```
